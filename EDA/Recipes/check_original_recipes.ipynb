```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Read in json file
df = pd.read_json('../../datasets/recipes_with_nutritional_info.json')
df.head()
```

```python
# title = "Dr. Oz. Energizing Breakfast: Canned Pumpkin Oatmeal"
# title = "Zizania Chicken Salad"
# title = "Hawaiian Chicken Stir-Fry"
# title = "Easy Balsamic-Honey-Mustard Glazed Salmon"
title = "Crunchy Broccoli Salad"
# Filter the dataframe for the specific recipe title
recipe_df = df[df['title'] == title]
recipe_df = recipe_df.reset_index(drop=True)
recipe_df
```

```python
# join quantity, unit, and ingredients into a list of combined strings
# recipe_df['ingredients_combined'] = recipe_df.apply(lambda row: f"{row['quantity']} {row['unit']} {row['ingredients']}", axis=1)
```

```python
print("Ingredients:")
for i in range(len(recipe_df['quantity'].values[0])):
    print(f"{recipe_df['quantity'].values[0][i]['text']} {recipe_df['unit'].values[0][i]['text']} {recipe_df['ingredients'].values[0][i]['text']}")
print("\nInstructions:")
for i in range(len(recipe_df['instructions'].values[0])):
    print(f"Step {i+1}: {recipe_df['instructions'].values[0][i]['text']}")
```

```python
# save recipe_df to a json file
recipe_df.to_json('../../datasets/dr_oz_recipe.json', orient='records', lines=True)
```
