```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Read in json file
# df = pd.read_json('../../datasets/recipes_with_nutritional_info.json')
df = pd.read_json('../../datasets/recipes_with_servings_and_normalized_nutrients.json', lines=True)
# df_corr = pd.read_json('../../datasets/recipes_with_servings_and_normalized_nutrients_corrected_quantities.json', lines=True)
df_corr = pd.read_json("../../VectorDB/recipes-vdb/data/llama3_1_vllm_async_corrected_quantities.jsonl", lines=True)
df_corr.head()
```

```python
title = "Dr. Oz. Energizing Breakfast: Canned Pumpkin Oatmeal"
# title = "Zizania Chicken Salad"
# title = "Hawaiian Chicken Stir-Fry"
# title = "Easy Balsamic-Honey-Mustard Glazed Salmon"
# title = "Crunchy Broccoli Salad"
# Filter the dataframe for the specific recipe title
# Get the row index of the recipe with the given title

row_index = df.index[df['title'] == title][0]
recipe_df = df[df['title'] == title]
recipe_df = recipe_df.reset_index(drop=True)
recipe_df
```

```python
# get the row index of df corresponding to the recipe title

row_index = recipe_df.index[0]
row_index
```

```python
row_index = df_corr.index[df_corr['title'] == title][0]
print(row_index)
recipe_df_corr = df_corr[df_corr['title'] == title]
recipe_df_corr = recipe_df_corr.reset_index(drop=True)
recipe_df_corr
```

```python
# join quantity, unit, and ingredients into a list of combined strings
# recipe_df['ingredients_combined'] = recipe_df.apply(lambda row: f"{row['quantity']} {row['unit']} {row['ingredients']}", axis=1)
```

```python
print("Ingredients:")
for i in range(len(recipe_df['quantity'].values[0])):
    print(f"{recipe_df['quantity'].values[0][i]['text']} {recipe_df['unit'].values[0][i]['text']} {recipe_df['ingredients'].values[0][i]['text']}")
print("\nInstructions:")
for i in range(len(recipe_df['instructions'].values[0])):
    print(f"Step {i+1}: {recipe_df['instructions'].values[0][i]['text']}")
print("\nMacros:")
# Loop through macro nutrients and sum up per serving values
nutr_totals = {}
for nutr in recipe_df['nutr_per_ingredient'].values[0]:
    for k, v in nutr.items():
        # print(k,v)
        nutr_totals[k] = nutr_totals.get(k, 0) + v
n_servings = recipe_df['n_servings'].values[0]
# Print out macro nutrient values
for key, val in nutr_totals.items():
    print(f"{key}: {val/n_servings}")

# macro_key_vals = {
#     "total_fat_per_serving": "Fat (g)",
#     "total_pro_per_serving": "Protein (g)",
#     "total_sug_per_serving": "Carbs (g)",
#     "total_nrg_per_serving": "Calories"}
# for key, val in macro_key_vals.items():
#     print(f"{val}: {recipe_df[key].values[0]}")
```

```python
print("Ingredients:")
for i in range(len(recipe_df_corr['corrected_quantity'].values[0])):
    print(f"{recipe_df_corr['corrected_quantity'].values[0][i]['text']} {recipe_df_corr['unit'].values[0][i]['text']} {recipe_df_corr['ingredients'].values[0][i]['text']}")
print("\nInstructions:")
for i in range(len(recipe_df_corr['instructions'].values[0])):
    print(f"Step {i+1}: {recipe_df_corr['instructions'].values[0][i]['text']}")
macro_key_vals = {
    "total_fat_per_serving": "Fat (g)",
    "total_pro_per_serving": "Protein (g)",
    "total_sug_per_serving": "Carbs (g)",
    "total_nrg_per_serving": "Calories"}

print("\nMacros:")
for key, val in macro_key_vals.items():
    print(f"{val}: {recipe_df_corr[key].values[0]}")
```
